```python
df = pd.read_csv('/content/Social_Network_Ads.csv')
df.head()
```

```python
print(df.shape)

# Converting Male/Female strings to 0/1
gender = {'Male': 0,'Female': 1}
df['Gender'] = [gender[item] for item in df['Gender']]

x = df[['User ID', 'Gender', 'Age', 'EstimatedSalary']]

print(x)
y = df['Purchased']
```

```python
# Logistic Regression implementation from scratch
class MyLogisticRegression:
    def __init__(self, alpha=0.01, iterations=100000, fit_intercept=True):
        self.alpha = alpha  #Learning Rate
        self.iterations = iterations #Finite Iterations for algorithm convergence
        self.fit_intercept = fit_intercept  #Set if to true if intercept needed 

    def _conc_intercept(self, X):  # Append Intercept 
        intercept = np.ones((X.shape[0], 1))  
        return np.concatenate((intercept, X), axis=1) 

    def _logistic(self, z): # Logistic(Sigmoid) Function for Logistic Regression
        return 1 / (1 + np.exp(-z))

    def _hypothesis(self, h, y): # Binary Cross Entropy Function
        return (-y * np.log(h) - (1 - y) * np.log(1 - h)).mean()

    def fit(self, X, y):  # Using Gradient Descent to calculate minima of the function
        if self.fit_intercept:
            X = self._conc_intercept(X) 

        self._features = np.zeros(X.shape[1]) 
        for i in range(self.iterations):  
            z = np.dot(X, self._features) 
            h = self._logistic(z)
            gradient = np.dot(X.T, (h - y)) / y.size
            self._features -= self.alpha * gradient
        
    
    def predict(self, X, threshold = 0.5):  # Prediction Function if prob >= thresh then one class else another class
        if self.fit_intercept:
            X = self._conc_intercept(X)
        return self._logistic(np.dot(X, self._features)) >= threshold
```

```python
plt.figure(figsize = (8,8))
ax = plt.axes(projection ='3d')

df0 = df[df.Purchased == 0]
df1 = df[df.Purchased == 1]

# Classifying Purchased as blue and not purchased as red data points in form of a scatter plot 
ax.scatter(df0['Gender'].values, df0['Age'].values,df0['EstimatedSalary'].values, c = 'red')
ax.scatter(df1['Gender'].values, df1['Age'].values,df1['EstimatedSalary'].values, c = 'blue')
plt.legend(['Not Purchased', 'Purchased'])
ax.set_title('Scatter plot of Features')
ax.set_xlabel('Gender')
ax.set_ylabel('Age')
ax.set_zlabel('EstimatedSalary')
```

```python
regressor = MyLogisticRegression()
X = x
# print(X,y)

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.1, random_state=4)
regressor.fit(X_train, y_train)

y_pred = regressor.predict(X_test)
score = accuracy_score(y_pred, y_test)
print(score)

confusion_mat = confusion_matrix(y_test, y_pred)

# print(y_predicted)
graph_cm_matrix = pd.DataFrame(confusion_mat, columns=['Actual True:1', 'Actual False:0'], 
                                 index=['Predicted Positive:1', 'Predicted Negative:0'])

sns.heatmap(graph_cm_matrix, annot=True, fmt='d')
```

```python
# Function to calculate Classification metrics from the confusion matrix
def calculate_classification_metrics(confusion_mat):
   TP = confusion_mat[0,0]
   FP = confusion_mat[0,1]
   FN = confusion_mat[1,0]
   TN = confusion_mat[1,1]
   accuracy = (TP + TN)/(TP + TN + FP + FN)
   precision = (TP)/(TP + FP)
   recall = (TP)/(TP + FN)
   F1 = (2 * precision * recall)/(precision + recall)
   print(f'Accuracy: {accuracy}')
   print(f'Precision: {precision}')
   print(f'Recall: {recall}')
   print(f'F1 Score: {F1}')
```

```python
calculate_classification_metrics(confusion_mat) 
```

***Question 5: Imbalanced dataset typically refers to a dataset where the classes are not represented
equally. Classification problems having multiple classes with imbalanced dataset
present a different challenge than a binary classification problem. The skewed

***20 News groups dataset. Since this dataset is a balanced one, you will perform the pre-
processing to create an imbalanced version of the dataset (by removing some news
articles from some groups). One example is given below. Perform multiclass
classification using logistic regression on both the balanced and the imbalanced version
of the dataset. Compare the performance in each case by obtaining the confusion matrix

You can refer to this article for a better understanding of multiclass classification using
logistic regression.***

```python
# We use Iris Dataset first of all
iris_df = pd.read_csv('/content/Iris.csv')
print(iris_df.head())
```

```python
# Picking Unique Species among the Species column
spec = iris_df['Species'].unique()

# Splitting the dataframe into small dataframes having data from the same species
iris_df0 = iris_df[iris_df['Species'] == spec[0]] 
iris_df1 = iris_df[iris_df['Species'] == spec[1]] 
iris_df2 = iris_df[iris_df['Species'] == spec[2]] 
```

***Balanced Dataset***

```python
# Creating Balanced Dataset from the Given Iris dataset by selecting a fraction of the rows from the splitted dataframes
new_subset_iris_df0 = iris_df0.sample(frac = 0.8) 
new_subset_iris_df1 = iris_df1.sample(frac = 0.8) 
new_subset_iris_df2 = iris_df2.sample(frac = 0.8)
```

```python
new_subset = new_subset_iris_df0.append(new_subset_iris_df1).append(new_subset_iris_df2)
cols = new_subset.columns[1:-1]
species = {'Iris-setosa': 0,'Iris-versicolor': 1, 'Iris-virginica' : 2}
new_subset['Species'] = [species[item] for item in new_subset['Species']]
X = new_subset[cols]
y = new_subset[new_subset.columns[-1]]
```

```python
clf = LogisticRegression(max_iter=10000)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=4)
clf.fit(X_train, y_train)

pred_y = clf.predict(X_test)
# print(pred_y)
conf_matrix = confusion_matrix(y_test, pred_y)
# print(conf_matrix)
graph_cm_matrix = pd.DataFrame(conf_matrix, columns=['Iris-setosa: 0','Iris-versicolor: 1', 'Iris-virginica : 2'], 
                                 index=['Iris-setosa: 0','Iris-versicolor: 1', 'Iris-virginica : 2'])

sns.heatmap(graph_cm_matrix, annot=True, fmt='d')
```

```python
# Classification Report
print(classification_report(y_test, pred_y))

print('Macro')
macro = precision_recall_fscore_support(y_test, pred_y, average='macro')
print(f'Precision = {macro[0]}\nRecall = {macro[1]}\nF1Score = {macro[2]}')

print('\nMicro')
micro = precision_recall_fscore_support(y_test, pred_y, average='micro')
print(f'Precision = {micro[0]}\nRecall = {micro[1]}\nF1Score = {micro[2]}')

print('\nWeighted')
weighted = precision_recall_fscore_support(y_test, pred_y, average='weighted')
print(f'Precision = {weighted[0]}\nRecall = {weighted[1]}\nF1Score = {weighted[2]}')
```
